##### 4. Load the data into a list of lists

```python
import csv
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

```python
sat_scores = []
with open('../assets/sat_scores.csv', 'rU') as data:
    scores = csv.reader(data)
    for i in scores:
        sat_scores.append(i)
```

```python
sat_scores_data = sat_scores[1:]
```

##### 7. Create a list of State names extracted from the data. (Hint: use the list of labels to index on the State column)

```python
state_list = [i[0:1] for i in sat_scores_data]
scores_list = [i[1:4] for i in sat_scores_data]
rates_list = [[i[1]] for i in sat_scores_data]
verbal_list = [[i[2]] for i in sat_scores_data]
math_list = [[i[3]] for i in sat_scores_data]
```

##### 8. Print the types of each column

```python
print(type(state_list))
print(type(rates_list))
print(type(verbal_list))
print(type(math_list))
```

##### 9. Do any types need to be reassigned? If so, go ahead and do it.

```python
data_convert = []
for row in sat_scores_data:
    convert = []
    for i, col in enumerate(row):
        if i == 0:
            convert.append(col)
        else:
            convert.append(float(col))

    data_convert.append(convert)
```

```python
data_convert
```

```python
# individual list for rate
rate_int = [int(index) for x in rates_list for index in x]
```

```python
# individual list for verbal
verbal_int = [int(index) for x in verbal_list for index in x]
```

```python
# individual list for math
math_int = [int(index) for x in math_list for index in x]
```

##### 10. Create a dictionary for each column mapping the State to its respective value for that column. 

```python
# each state in state_data append row of sat_scores
state_data = {}
for column_name in data_convert:
    states = column_name[0]    
    data = column_name[1:4]
    state_data[states] = data

```
